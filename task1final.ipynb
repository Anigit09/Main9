```python
import cv2 as cv
import pandas as pd
import matplotlib.pyplot as plt
img=cv.imread("robot.png")
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
e1=cv.resize(img,None,fx=0.2,fy=0.2,interpolation=cv.INTER_AREA)
e2=cv.resize(img,None,fx=0.2,fy=0.2,interpolation=cv.INTER_AREA)
print(e1.shape)
print(e2.shape)
```

```python
plt.imshow(e1)
```

```python
e1=e1[0:212,50:200]
e2=e2[0:212,50:200]
```

```python
img[650:862,30:180]=e2
img[650:862,1000:1150]=e1
```

```python
plt.imshow(img)
plt.show()
```

```python
red=cv.cvtColor(e1,cv.COLOR_BGR2RGB)
gray=cv.cvtColor(red,cv.COLOR_RGB2GRAY)
gray1=cv.cvtColor(gray,cv.COLOR_GRAY2BGR)
```

```python
img[650:862,30:180]=gray1
img[650:862,1000:1150]=red
```

```python
plt.imshow(img)
plt.show()
```
